```python
import pandas as pd
data={
    "CustomerID":[101,102,103,104,105],
    "Age":[23,45,35,52,40],
    "Gender":["Male","Female","Female","Male","Female"],
    "Annual_Income":[30000,60000,50000,80000,70000],
    "Spending_Score":[39,81,6,77,40]
}
```

```python
#create variable Income_per_Age by dividing Annual_income by Age
#create a variable Monthly_Income
#Find unique values in Gender column
#Find frequency counts of gender column
#create dummy variables for gender without dropping any column
#create dummy variables for gender with drop_first=True
#encode gender using map
#encode gender using replace
#create age_group column(Young,Middle-ages,Senoir)
```

```python
df = pd.DataFrame(data)
df["Income_per_Age"] = df["Annual_Income"] / df["Age"]
print(df)
```

```python
df["Monthly_Income"] = df["Annual_Income"] / 12
print(df)
```

```python
unique_genders = df["Gender"].unique()
print("Unique values in Gender:", unique_genders)
```

```python
gender_counts = df["Gender"].value_counts()
print("\nFrequency counts:\n", gender_counts)
```

```python
gender_dummies = pd.get_dummies(df["Gender"])
print("\nDummy variables without any drop):\n", gender_dummies)
```

```python
# dummy variables for Gender (drop_first=True)
gender_dummies_drop = pd.get_dummies(df["Gender"], drop_first=True)
print("\nDummy variables (drop_first=True):\n", gender_dummies_drop)
```

```python
# Encode Gender using map
df["Gender_map"] = df["Gender"].map({"Male": 0, "Female": 1})
df
```

```python
# Encode Gender using replace
df["Gender_replace"] = df["Gender"].replace({"Male": 1, "Female": 0})
df
```

```python
# Create Age_Group column (Young <30, Middle-aged 30–50, Senior >50)
def age_group(age):
    if age < 30:
        return "Young"
    elif 30 <= age <= 50:
        return "Middle-aged"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

print("\nFinal DataFrame:\n", df)
```

```python
# Bin age into 3 equal-width categories
df["Age_bin"] = pd.cut(df["Age"],
                        bins=3,
                          labels=["Young", "Middle-aged", "Senior"])
print(df[["Age", "Age_bin"]])
age_dummies = pd.get_dummies(df["Age_bin"], prefix="Age")

df = pd.concat([df, age_dummies], axis=1)

print("Final DataFrame:\n", df)
```

```python
#scale Annual_income using MaxAbs
print(df["Annual_Income"].describe(), "\n")
df["Annual_Income_scaled"] = df["Annual_Income"] / df["Annual_Income"].abs().max()

print(df[["Annual_Income", "Annual_Income_scaled"]].head(), "\n")
print(df["Annual_Income_scaled"].describe())
```

```python
#minmax normalization
df['Spending_Score'].describe()
print('\n')
min_x=df['Spending_Score'] - df['Spending_Score'].min()
min_max=df['Spending_Score'].max()-df['Spending_Score'].min()
df['Spending_Score']=min_x/min_max
print(df['Spending_Score'].head())
print("\n")
print(df['Spending_Score'].describe())
```
