```python
import numpy as np
import pandas as pd
data=pd.read_csv('student_dataset.csv')
print(data.info())
print(data.describe())
```

```python
print(data.columns)
```

```python
data.sort_values(by=['Gender','BloodType'], ascending=True, inplace=False)
```

```python
print(data.groupby(['Gender','BloodType'])['Height'].max())
```

```python
max_height=data['Height'].max()
max_height
```

```python
my_dict={"Size":["L","L","M","M","M","S","S","S","S"],
         "Type":["A","A","A","B","B","A","A","B","B"],
         "Location":["L1","L1","L1","L2","L2","L1","L2","L2","L1"],
         "A":[1,2,2,3,3,4,5,6,7],
         "B":[2,4,5,5,6,6,8,9,9]}
df=pd.DataFrame(my_dict)
df
```

```python
dfr=pd.pivot_table(df,index=['Size','Type'],columns='Location',values=['A','B'],fill_value=0)
dfr
```

```python
pd.pivot_table(df,index=['Size','Type'],values=['A','B'],aggfunc=np.median)
```

```python
df.groupby(['Size','Type'])[['A','B']].mean()
```

```python
pd.pivot_table(df,index=['Size','Type'],values=['A','B'],aggfunc={'A':np.max,'B':np.min})
```
